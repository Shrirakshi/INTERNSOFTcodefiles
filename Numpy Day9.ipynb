```python
import time
import numpy as np
M=32
N=64
```

```python
x=np.linspace(0.0,1.0,M,endpoint=False)
y=np.linspace(0.0,1.0,N,endpoint=False)
```

```python
t0=time.time()
```

```python
g=np.zeros((M,N))
```

```python
for i in range(M):
    for j in range(N):
        g[i,j]=np.sin(2.0*np.pi*x[i]*y[j])
```

```python
t1=time.time()
print("time elapsed:{} s".format(t1-t0))
```

## Concatenation of arrays 

```python
x=np.array([1,2,3,4])
y=np.array([5,4,3,2,1])
np.concatenate([y,x])
```

```python
z=[9,4,6]
print(np.array([x,y,z]))
```

```python
grid=np.array([[50,60,70],
             [80,90,100]])
```

```python
#concatenate along the first axis
np.concatenate([grid,grid])
```

```python
#conctenate along second axis(zero-indexed)
np.concatenate([grid,grid],axis=1)
```

## Splitting of arrays

```python
x=[1,2,3,4,5,5,4,3,2,1,9,4,6]
x1,x2,x3,x4=np.split(x,[3,5,8])
print(x1,x2,x3,x4)
```

```python
grid=np.arange(20).reshape((5,4))
grid
```

```python
left,right=np.hsplit(grid,[2])
print(left)
print(right)
```
